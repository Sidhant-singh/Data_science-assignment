1.DataFrame.size:

DataFrame.size returns the total number of elements in the DataFrame. It calculates the product of the number of rows and columns.
The result is a single integer representing the total number of elements in the DataFrame.

2.DataFrame.shape:

DataFrame.shape returns a tuple representing the dimensions of the DataFrame. The tuple contains two elements: the number of rows and the number of columns.
The result is a tuple, and you can access the number of rows using shape[0] and the number of columns using shape[1].

```python
# Q7
import pandas as pd

df = pd.DataFrame({
    'Email': ['john.doe@example.com', 'siddhant@example.com', 'pwskills@example.com']
})

def username(email):
    st = ""
    i = 0
    while email[i] != '@':
        st += email[i]
        i += 1
    return st

df['Username'] = df['Email'].apply(username)
print(df)
```

```python
#Q8
def filter_dataframe(df):
    # Select rows where 'A' is greater than 5 and 'B' is less than 10
    selected_rows = df[(df['A'] > 5) & (df['B'] < 10)]
    
    # Create a new DataFrame with the selected rows
    result_df = pd.DataFrame(selected_rows)
    
    return result_df

data = {'A': [3, 8, 6, 2, 9],
        'B': [5, 2, 9, 3, 1],
        'C': [1, 7, 4, 5, 2]}

df = pd.DataFrame(data)

result_dataframe = filter_dataframe(df)

print(result_dataframe)
```

```python
#Q9
def calculate_stats(df):
    # Extract the 'Values' column
    values_column = df['Values']
    
    # Calculate mean, median, and standard deviation
    mean_value = values_column.mean()
    median_value = values_column.median()
    std_deviation = values_column.std()
    
    # Create a dictionary to store the results
    result_dict = {
        'Mean': mean_value,
        'Median': median_value,
        'Standard Deviation': std_deviation
    }
    
    return result_dict
```

```python
#Q10
def add_moving_average(df):
    # Ensure the 'Date' column is in datetime format
    df['Date'] = pd.to_datetime(df['Date'])
    
    # Sort the DataFrame by 'Date' if it's not already sorted
    df = df.sort_values(by='Date')
    
    # Calculate the moving average with a window of size 7
    df['MovingAverage'] = df['Sales'].rolling(window=7, min_periods=1).mean()
    
    return df
```

```python
#Q11

def add_weekday_column(df):
    # Ensure the 'Date' column is in datetime format
    df['Date'] = pd.to_datetime(df['Date'])
    
    # Extract weekday name and create the 'Weekday' column
    df['Weekday'] = df['Date'].dt.strftime('%A')
    # %A is a format code for weekdays.
    
    return df

# Example DataFrame
data = {'Date': ['2023-01-01', '2023-01-02', '2023-01-03', '2023-01-04', '2023-01-05']}

df = pd.DataFrame(data)

df_with_weekday = add_weekday_column(df)

print(df_with_weekday)
```
