
Q1: What is Matplotlib? Why is it used? Name five plots that can be plotted using the Pyplot module of Matplotlib ?

Matplotlib is a popular 2D plotting library for Python. It provides a flexible platform for creating static, animated, and interactive visualizations in Python.
Five plots that can be used :
1.Line plot
2.Scatter plot
3.Bar plot
4.Histogram
5.pie chart

```python
##Q2
import numpy as np
import matplotlib.pyplot as plt
np.random.seed(3)
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))
plt.scatter(x,y)
```

Q3
The subplot() function in Matplotlib is used to create multiple plots in the same figure. It allows you to arrange and display several plots within a single figure, making it easier to compare and analyze different datasets or aspects of your data.

```python
##Q3 
x = np.array([0, 1, 2, 3, 4, 5])
y1 = np.array([0, 100, 200, 300, 400, 500])
y2 = np.array([50, 20, 40, 20, 60, 70])
y3 = np.array([10, 20, 30, 40, 50, 60])
y4 = np.array([200, 350, 250, 550, 450, 150])
plt.subplot(2,2,1)
plt.plot(x,y1)
plt.title("plot 1")
```

```python
plt.subplot(2,2,2)
plt.plot(x,y2)
plt.title("plot 2")
```

```python
plt.subplot(2,2,3)
plt.plot(x,y3)
plt.title("plot 3")
```

```python
plt.subplot(2,2,4)
plt.plot(x,y4)
plt.title("plot 4")
```

A box plot (or box-and-whisker plot) is a statistical visualization that provides a summary of the distribution of a dataset. It displays the central tendency, spread, and potential outliers of the data.

```python
box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)
data = [box1,box2]
plt.boxplot(data,labels=['Box 1','Box 2'])
plt.title('Box Plot of box1 and box2')
plt.xlabel('Box')
plt.ylabel('Values')
plt.show()
```
